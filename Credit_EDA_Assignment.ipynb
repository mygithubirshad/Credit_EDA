# Feature slection
## Data cleaning on appications data.

```python
# checking app shape
app.shape
```

```python
# checking null values

missing_values = pd.DataFrame(app.isnull().sum().sort_values()).reset_index()
missing_values.rename(columns = {'index' : 'Columns', 0 : 'Null_Values'},inplace = True)
missing_values
```

```python
# converting null values to percentage and storing in a new columns.
missing_values["miss_per"] = missing_values['Null_Values']/app.shape[0]*100
missing_values
```

```python
# filtering the null values >=40 and droping the null values >= 40.
miss_col  = missing_values[missing_values['miss_per'] >= 40]['Columns'].to_list()

app_msing_rmd = app.drop(labels= miss_col, axis = 1)
app_msing_rmd.shape
```

```python
app_msing_rmd.head()
```

```python
# checking the data types on columns.
app_msing_rmd.info()
```

```python
#checking the unique values.
app_msing_rmd.nunique().sort_values()
```

## Data cleaning on Previous_Applications data.

```python
# checking pre shape
pre.shape
```

```python
# checking null values

pre_missing_values = pd.DataFrame(pre.isnull().sum().sort_values()).reset_index()
pre_missing_values.rename(columns = {'index' : 'Columns', 0 : 'Null_Values'},inplace = True)
pre_missing_values
```

```python
# converting null values to percentage and storing in a new columns.
pre_missing_values["miss_per"] = pre_missing_values['Null_Values']/pre.shape[0]*100
pre_missing_values
```

```python
# filtering the null values >=40 and droping the null values >= 40.
pre_miss_col  = pre_missing_values[pre_missing_values['miss_per'] >= 40]['Columns'].to_list()

pre_msing_rmd = pre.drop(labels= pre_miss_col, axis = 1)
pre_msing_rmd.shape
```

```python
# checking the data types on columns.
pre_msing_rmd.info()
```

```python
#checking the unique values.
pre_msing_rmd.nunique().sort_values()
```

# Imputing the missing values in Applications data

```python
# checking days columns
days_col = []
for col in app_msing_rmd.columns:
    if col.startswith("DAYS"):
        days_col.append(col)
days_col        
```

```python
# using absalute to convert -ve to +ve.
for col in days_col:
    app_msing_rmd[col] = abs(app_msing_rmd[col])
    
```

```python
# checking head() to confirm the change.
app_msing_rmd.head()
```

```python
# changing the data types
app_msing_rmd['DAYS_REGISTRATION'] = app_msing_rmd['DAYS_REGISTRATION'].astype(int, errors ='ignore')
app_msing_rmd['CNT_FAM_MEMBERS'] = app_msing_rmd['CNT_FAM_MEMBERS'].astype(int, errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_HOUR'] = app_msing_rmd['AMT_REQ_CREDIT_BUREAU_HOUR'].astype(int , errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_DAY'] = app_msing_rmd['AMT_REQ_CREDIT_BUREAU_DAY'].astype(int , errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_WEEK'] =app_msing_rmd['AMT_REQ_CREDIT_BUREAU_WEEK'].astype(int , errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_MON'] = app_msing_rmd['AMT_REQ_CREDIT_BUREAU_MON'].astype(int , errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_QRT'] = app_msing_rmd['AMT_REQ_CREDIT_BUREAU_QRT'].astype(int , errors ='ignore')
app_msing_rmd['AMT_REQ_CREDIT_BUREAU_YEAR'] = app_msing_rmd['AMT_REQ_CREDIT_BUREAU_YEAR'].astype(int , errors ='ignore')
app_msing_rmd['OBS_30_CNT_SOCIAL_CIRCLE'] = app_msing_rmd['OBS_30_CNT_SOCIAL_CIRCLE'].astype(int , errors = 'ignore')
app_msing_rmd['DEF_30_CNT_SOCIAL_CIRCLE'] = app_msing_rmd['DEF_30_CNT_SOCIAL_CIRCLE'].astype(int, errors ='ignore')
app_msing_rmd['DEF_60_CNT_SOCIAL_CIRCLE'] = app_msing_rmd['DEF_60_CNT_SOCIAL_CIRCLE'].astype(int, errors ='ignore')
app_msing_rmd['DEF_60_CNT_SOCIAL_CIRCLE'] = app_msing_rmd['DEF_60_CNT_SOCIAL_CIRCLE'].astype(int, errors ='ignore')
```

```python
# droping unwanted columns.
unwanted = ["FLAG_DOCUMENT_19","FLAG_DOCUMENT_20","FLAG_DOCUMENT_21", "FLAG_DOCUMENT_9"  ,"FLAG_DOCUMENT_17"]
app_msing_rmd.drop(labels = unwanted ,axis = 1 , inplace = True)
            
```

#  Merging the two datasets, i.e. application_dataset and previous_application

```python
plt.figure(figsize=(15,5))


sns.countplot(data = merg, x="NAME_CASH_LOAN_PURPOSE" ,hue="NAME_CONTRACT_STATUS" )
plt.xticks (rotation = 90)
plt.yscale("log")

```
